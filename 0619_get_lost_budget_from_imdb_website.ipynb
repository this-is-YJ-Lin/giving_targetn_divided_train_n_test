```python
# -*- coding: utf-8 -*-
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup

HTML_PARSER   = "html.parser"
USER_AGENT    = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/59.0.3071.115 Safari/537.36'
HEADERS       = {'User-Agent': USER_AGENT}
MOVIE_ID_URL  = 'https://www.imdb.com/title/'
head = ["imdb_id", "budget"]
fileName = '100_5093_lost_budget 拷貝.csv'

def title(file):
	count = 0
	df = pd.read_csv(file)
	for i in range(4020, 5864):
		print(i, df.imdb_id[i])
		data_list = Movie(df.imdb_id[i])
		df_write = pd.DataFrame(data_list , columns = head)
		df_write.to_csv("budget4020_to_5864.csv", mode='a', header=False)
		print('count: ',count)
		print(i,"write ok")
		count += 1

def Movie(title):
	data_list = []
	movie_req = requests.get(MOVIE_ID_URL+title, headers = HEADERS)
	if movie_req.status_code == requests.codes.ok:
		print('URL OK')
		soup = BeautifulSoup(movie_req.content, HTML_PARSER)
		if len(soup.select('div.txt-block')) == 0:
			budget = 'no data'
		else:
			for i in range(0, len(soup.select('div.txt-block'))):
				if 'Budget:' in soup.select('div.txt-block')[i].text:
					budget = soup.select('div.txt-block')[i].text.strip().replace('Budget:', '')
					budget = budget.replace('(estimated)','').strip()
					break
				else:
					budget = 'no data'
			print(budget)
	data_list.append([title, budget])
	return data_list

if __name__ == '__main__':
	start = time.time()
	title(fileName)
	print('All cost',time.time()-start)
```
