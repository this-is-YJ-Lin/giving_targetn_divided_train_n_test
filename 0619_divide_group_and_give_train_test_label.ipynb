```python
df_10038 = pd.read_csv('all_input_prepared_3436_movies.csv')
df_10038.head()
```

```python
df_10038['year_box_office_label'] = 0
df_10038['train_test'] = 'N/A'
df_10038.head()
```

```python
group_by_year = df_10038.groupby('year')
```

```python
group_by_year.size()
```

```python
group_by_year.get_group(2000)
```

```python
total_range = 10

def label_by_year_sort_woth_box_office(year):
    
    group_year_movie = group_by_year.get_group(year)
    each_group_num = int(len(group_year_movie)/total_range)
    label_count = 0

    for i in range(0, len(group_by_year.get_group(year)),each_group_num):
        #group_year_movie.year_box_office_label.iloc[ i : i+each_group_num ] = label_count
        df_10038.year_box_office_label[group_by_year.groups[year][ i : i+each_group_num ]] = label_count
        label_count = label_count+1
        #print(df_10038.year_box_office_label[group_by_year.groups[year][i]])
        #if df_10038.year_box_office_label[group_by_year.groups[year][i]] < 0:
            #print(df_10038.year_box_office_label[group_by_year.groups[year][i]])
            #df_10038.year_box_office_label[group_by_year.groups[year][ i : i+each_group_num ]] = 0
            
for i in group_by_year.groups.keys():
    label_by_year_sort_woth_box_office(i)

for k in range(0, len(df_10038.year_box_office_label)):
    if df_10038.year_box_office_label[k] > total_range-1:
        df_10038.year_box_office_label[k] = total_range-1
```

```python
df_sort_y_Rnum = df_10038.groupby(['year_box_office_label'])
df_sort_y_Rnum.size()
```

```python
#len(index_num_list)
for sp_keys in df_sort_y_Rnum.groups.keys():
    print('sp_key: ',sp_keys)
    spcified_year_df = df_sort_y_Rnum.get_group(sp_keys)
    threshold = int(len(spcified_year_df)*0.2)
    print(threshold)
    for index_i in range(0,len(df_sort_y_Rnum.groups[sp_keys])):
        if index_i <= threshold:
            #print('train')
            #print(df_sort_y_Rnum.groups[sp_keys][index_i])
            df_10038['train_test'][ df_sort_y_Rnum.groups[sp_keys][index_i] ] = 'test'
        else:
            #print('test')
            #print(df_sort_y_Rnum.groups[sp_keys][index_i])
            df_10038['train_test'][ df_sort_y_Rnum.groups[sp_keys][index_i] ] = 'train'
```
